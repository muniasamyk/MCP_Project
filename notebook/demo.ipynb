```python
# Example 1: Get employees from a department
print("EXAMPLE 1: Fetch Employees by Department\n")

# The user's question
query = "Fetch employee details where department is AI"
print(f"Query: {query}\n")

# Step 1: Planner reads the query
print("Step 1: Planner Agent")
plan = planner_agent(query)
print(f"Plan: {json.dumps(plan, indent=2)}\n")

# Step 2: Executor runs the query
print("Step 2: Executor Agent")
result = executor_agent(plan)

# Show the results
if result.get('status') == 'success':
    rows = result.get('row_count')
    for row in result.get('data', []):
        print(f"  {row}")
else:
    print(f"Error: {result.get('message')}")
```
